## Preliminaries

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import gridspec
import pandas_datareader.data as web

# Show plots in notebook
%matplotlib inline
```

## Gather data and derive the daily return

Using the pandas datareader, we retrieve information on Ford's stock price and calculate a daily return as:

$ X_t = ln(\frac{CP_t}{CP_{t-1}} ) $, 

where, $X_t $ is the daily return, $CP_t $ is the daily closing price, and $CP_{t-1}$ is yesterday's closing price.  The source of the data is Yahoo! Finance.

```python
# Gather data from Yahoo! Finance for Ford's daily stock price since 2013
F = web.DataReader(name='F', data_source='yahoo', start='2013-1-1')
# Calculate the daily return
F['Return'] = (np.log(F['Close'] / F['Close'].shift(1)) * 100)
```

## Calculate mean annual prices for report

To calculate annual ratios, let's take a look at the average annual stock price of both Ford and GM

```python
# Set index for each year
F2013 = F['Close'][0:252]
F2014 = F['Close'][252:504]
F2015 = F['Close'][504:756]
F2016 = F['Close'][756:]

print 'Ford 2013: $' + str(np.round(np.mean(F2013), 2))
print 'Ford 2014: $' + str(np.round(np.mean(F2014), 2))
print 'Ford 2015: $' + str(np.round(np.mean(F2015), 2))
print 'Ford 2016 to date: $' + str(np.round(np.mean(F2016), 2))
```

```python
#Gather GM stock price data
GM = web.DataReader(name='GM', data_source='yahoo', start='2013-1-1')

GM2013 = GM['Close'][0:252]
GM2014 = GM['Close'][252:504]
GM2015 = GM['Close'][504:756]
GM2016 = GM['Close'][756:]

print 'GM 2013: $' + str(np.round(np.mean(GM2013), 2))
print 'GM 2014: $' + str(np.round(np.mean(GM2014), 2))
print 'GM 2015: $' + str(np.round(np.mean(GM2015), 2))
print 'GM 2016 to date: $' + str(np.round(np.mean(GM2016), 2))
```

## Distribution and volatility of returns

Beyond averages, we also want to know about periods of exceptional loss. In other words, how volatile are the returns?

```python
# Name ranges by year in returns series
FR = F['Return'].dropna()
F2013R = F['Return'][0:252].dropna()
F2014R = F['Return'][252:504].dropna()
F2015R = F['Return'][504:756].dropna()
F2016R = F['Return'][756:].dropna()
```

```python
# Create histogram of daily returns
h = plt.hist(FR, normed=True, bins=50)

# print the mean, min, and standard deviation daily return
print "mean 2013-present: " + str( np.round(np.mean(FR),4) ) + "; " \
        + "min 2013-15: " + str( np.round(np.min(FR),2) )+ "; " \
        + "std dev 2013-15: " + str( np.round(np.std(FR),2) )

# and by year
print "mean 2013: " + str( np.round(np.mean(F2013R),4) ) + "; " \
        + "min 2013: " + str( np.round(np.min(F2013R),2) )+ "; " \
        + "std dev 2013: " + str( np.round(np.std(F2013R),2) )

print "mean 2014: " + str( np.round(np.mean(F2014R),4) ) + "; " \
        + "min 2014: " + str( np.round(np.min(F2014R),2) )+ "; " \
        + "std dev 2014: " + str( np.round(np.std(F2014R),2) )
        
print "mean 2015: " + str( np.round(np.mean(F2015R),4) ) + "; " \
        + "min 2015: " + str( np.round(np.min(F2015R),2) )+ "; " \
        + "std dev 2015: " + str( np.round(np.std(F2015R),2) )
        
print "mean 2016: " + str( np.round(np.mean(F2016R),4) ) + "; " \
        + "min 2016: " + str( np.round(np.min(F2016R),2) )+ "; " \
        + "std dev 2016: " + str( np.round(np.std(F2016R),2) )
```

We see above that ownership of Ford common stock has generated practically no return other than dividends since the beginning of 2013. However, during 2013, the average daily return was more than six percent. In 2014, it was 1.8 percent, and it was negative in both 2015 and so far in 2016 (-3.8 percent and -4.3 percent respectively). Volatility has been highest in the year to date period. In the three full years, 2013 had the highest volatility, as measured by standard deviation of daily return.

## Plotting the data

Now we can plot the data, including both the closing price and daily return, and noting what we've found above.

```python
#Create a figure
fig = plt.figure(figsize=(8, 5))

# Section it in two with the top one taller and tight spacing 
gs = gridspec.GridSpec(2, 1, height_ratios=[2, 1])
gs.update(wspace=0.1, hspace=0.1)

# Draw line plot of closing price in top figure
ax1 = plt.subplot(gs[0])
ax1.plot(F['Close'], label="Closing Price", color='darkblue')

# Draw a line plot of the daily return in the bottom figure
ax2 = plt.subplot(gs[1])
ax2.plot(F['Return'], label="Daily Return", color='dodgerblue')

# Set Figure labels
ax1.set_ylabel('USD')
ax1.set_xlabel('')
ax2.set_ylabel('Pct Change')

# Turn off the top figure x axis labels but leave ticks
ax1.axes.xaxis.set_ticklabels([])

# Adjust the x axis label size and rotation
plt.tick_params(axis='both', which='major', labelsize=9)
plt.setp( ax2.xaxis.get_majorticklabels(), rotation=20)

# Add annotation for 2013 drop
ann = ax1.annotate('Ford lowers profit estimate', ((F.index[238],16.35)), 
                  ((F.index[110],13.5)),
                  arrowprops=dict(arrowstyle="->",
                                 color='gray',
                                 patchB=None,
                                 shrinkB=0,
                                 connectionstyle="arc3,rad=-0.3"),
                  )

# Add annotation for 2014 drop
ann = ax1.annotate('Ford lowers profit estimate', ((F.index[432],16.2)),
                  ((F.index[110],13.5)),
                  arrowprops=dict(arrowstyle="->",
                                 color='gray',
                                 patchB=None,
                                 shrinkB=0,
                                 connectionstyle="arc3,rad=0.1"),
                  )

# Add annotation for 2014 below forecast
ann = ax1.annotate('Earnings below analyst forecast', ((F.index[661],14.7)),
                  ((F.index[360],12.8)),
                  arrowprops=dict(arrowstyle="->",
                                 color='gray',
                                 patchB=None,
                                 shrinkB=0,
                                 connectionstyle="arc3,rad=0.3"),
                  )

# Add annotation for profit disappointment
ann = ax1.annotate('Ford profit estimate disappoints', ((F.index[757],13.9)),
                  ((F.index[450],11.8)),
                  arrowprops=dict(arrowstyle="->",
                                 color='gray',
                                 patchB=None,
                                 shrinkB=0,
                                 connectionstyle="arc3,rad=0.2"),
                  )

# Set y axis limits for the returns plot, so that there is space for text.
plt.ylim(ymin=-12)
plt.ylim(ymax=11)

# Add the most recent close price and date
ax1.text(F.index[650], 16.8, str(F.index[-1].strftime('%Y-%m-%d'))\
         + ': $' + str(F.Close[-1]), fontsize=9)

# Add the daily return std dev by year
ax2.text(F.index[20], -10, 'Standard deviation by year: 2013: ' + \
         str( np.round(np.std(F2013R),2))+ ';   2014: ' + \
         str( np.round(np.std(F2014R),2))+ ';   2015: ' + \
         str( np.round(np.std(F2015R),2))+ ';   2016 to date: ' + \
         str( np.round(np.std(F2016R),2)), fontsize=9)


# add legends
ax1.legend(frameon=False, prop={'size':9})
ax2.legend(loc=1, frameon=False, prop={'size':9})

#Save as png
plt.savefig('ford.pdf');
```
