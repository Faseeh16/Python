```python
import requests
import pandas as pd
import config   ## File with API key

def write_txt(filename, filetext):
# Write label to txt file
    with open(filename, 'w') as text_file:
        text_file.write(filetext)

# Components of request
base = 'https://www.bea.gov/api/data/?&UserID={}&method=GetData'.format(config.bea_key)
url = '{}&datasetname={}&TableId={}&Frequency={}&Year={}&ResultFormat=json'.format(
    base, 'NIPA', 2, 'Q', 'LAST12')    # Adjust here as needed
df = pd.DataFrame(requests.get(url).json()['BEAAPI']['Results']['Data'])
df['date'] = pd.to_datetime(df['TimePeriod'])
codes = ['A191RL', 'DPCERY', 'A006RY', 'A019RY', 'A822RY']
df = df[df['SeriesCode'].isin(codes)].set_index(['date', 'SeriesCode'])['DataValue'].unstack(1)
df.to_csv('data/gdp_comp.csv')

# GDP in levels latest value
url = '{}&datasetname={}&TableId={}&Frequency={}&Year={}&ResultFormat=json'.format(
    base, 'NIPA', 6, 'Q', 'LAST1')    # Adjust here as needed
Y = pd.DataFrame(requests.get(url).json()['BEAAPI']['Results']['Data'])['DataValue'][0]

# Record strings for later use in chart label
date = '{} Q{}'.format(df.index[-1].year, df.index[-1].quarter)
ch = df['A191RL'][-1]  # GDP in growth
label = '{}: Real GDP: {}T; Growth: {}\%'.format(date, round(float(Y.replace(',', ''))/1000,1), ch)

write_txt('data/gdp_comp.txt', label) # Write chart label to txt
```
