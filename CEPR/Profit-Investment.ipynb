### Profit share of Net Domestic Product vs Investment Share

```python
import pandas as pd
import pandas_datareader.data as web
import datetime
%matplotlib inline
```

```python
series = ['CP', 'CPATAX', 'CPROFIT', 'PNFI', 'A362RC1Q027SBEA']
```

```python
# Retrieve data as pandas dataframe named df
df = web.DataReader(series, 'fred', datetime.datetime(1964,1,1))
```

```python
for s in series[:-1]:
    df[s+'_NDP'] = df[s] / df[series[-1]]
df
```

```python
#slist = [s + '_NDP' for s in series[:-1]]
slist = [s + '_NDP' for s in ['CPROFIT', 'CPATAX', 'PNFI']]
```

```python
df[slist].plot(figsize=(15, 5))
df[slist].pct_change(4).plot(figsize=(15, 5))
```

```python
df['CP_NDP'].pct_change(4).corr(df['PNFI_NDP'].pct_change(4))
```
